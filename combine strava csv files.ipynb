```python
import pandas as pd
import re
import glob
```

```python
path = r'data/'
all_files = glob.glob(path + "/*.csv")
r_workout = re.compile(r"workout.csv$")
r_workout_cadences = re.compile(r"workout_cadences.csv$")
r_workout_hrs = re.compile(r"workout_hrs.csv$")
r_workout_powers = re.compile(r"workout_powers.csv$")
r_workout_speeds = re.compile(r"workout_speeds.csv$")

r_indexes = re.compile(r"[0-9]+_[0-9]+")
strava_id_column = 'workout_strava_id'

for file in all_files:
    if r_workout.search(file):
        curr_indexes = r_indexes.search(file).group(0)
        curr_df = pd.read_csv(file)
        
        for f in all_files:
            found_indexes = r_indexes.search(f).group(0)
            if not f == file and found_indexes == curr_indexes:
                # append workout_strava_id column
                found_df = pd.read_csv(f)
                found_df[strava_id_column] = curr_df['workout_strava_id']
                found_df.to_csv(f, index=False)
```

```python
df_workout = pd.concat((pd.read_csv(f) for f in all_files if r_workout.search(f)))
df_workout_cadences = pd.concat((pd.read_csv(f) for f in all_files if r_workout_cadences.search(f)))
df_workout_hrs = pd.concat((pd.read_csv(f) for f in all_files if r_workout_hrs.search(f)))
df_workout_powers = pd.concat((pd.read_csv(f) for f in all_files if r_workout_powers.search(f)))
df_workout_speeds = pd.concat((pd.read_csv(f) for f in all_files if r_workout_speeds.search(f)))
```

```python
df_workout.to_csv("isn_workout.csv", encoding='utf-8', index=False)
df_workout_cadences.to_csv("isn_workout_cadences.csv", encoding='utf-8', index=False)
df_workout_hrs.to_csv("isn_workout_hrs.csv", encoding='utf-8', index=False)
df_workout_powers.to_csv("isn_workout_powers.csv", encoding='utf-8', index=False)
df_workout_speeds.to_csv("isn_workout_speeds.csv", encoding='utf-8', index=False)
```
